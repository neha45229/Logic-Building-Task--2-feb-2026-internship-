1.Unique words in a sentence

```python
from enum import unique
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = []
for word in words:
  if word not in unique_words:
    unique_words.append(word)
print("unique word count:",len(unique_words))
print("unique words:",unique_words)
```

2.finding highest employee with his name and salary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
for name in employees:
  if employees[name]==max(employees.values()):
    print("Highest Salary:",name,employees[name])
```

```python
numbers = [45, 22, 89, 10, 66]
print("List:",numbers)
max=numbers[0]
min=numbers[0]
for i in numbers:
  if i > max:
    max=i
  if i < min:
    min=i
print("max:",max)
print("min:",min)
```

4.count products above a price threshold

```python
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:
  if i > 1000:
    count+=1
print( "Products above 1000:",count)
```

5.Calculating attendence percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present_count=0
for i in attendance:
  if i == "P":
    present_count+=1
print("Attendance percentage:",(present_count/len(attendance))*100)


```
